```python
# importing the library
import numpy as np
import matplotlib.pyplot as plt

# data to be plotted
x = np.arange(1, 13)
y = np.array([2849, 1384, 911, 698, 560, 470, 414, 371, 344, 318, 306, 284])
 
# plotting
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)

plt.title("Total calculation time")
plt.xlabel("Number of threads")
plt.ylabel("Time [ms]")
plt.plot(x, y, color ="green", marker = "o")
plt.show()
```

```python
# importing the library
import numpy as np
import matplotlib.pyplot as plt

# data to be plotted
x = np.arange(2, 13)
y = np.array([1384, 911, 698, 560, 470, 414, 371, 344, 318, 306, 284])
y = 2849 / y

# plotting
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)

plt.title("Speedup quotient")
plt.xlabel("Number of threads")
plt.ylabel("Ts / Tn")
plt.plot(x, y, color ="red", marker = "o")
plt.show()
```
